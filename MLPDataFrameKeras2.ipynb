```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
```

```python
data_set = pd.read_csv('Iris.csv')
```

```python
#reemplazar las etiquetas string a int, ejemplo (Iris-cetosa=0)
le = LabelEncoder().fit(data_set.Species)
data_set.Species = le.transform(data_set.Species)
```

```python
#ver el dataframe despues de reemplazar las etiquetas
data_set.head()
```

```python
num_classes = 3
```

```python
model = Sequential()
model.add(Dense(100, input_shape=(4,), activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(data_set.iloc[:, 1:5], data_set.Species, epochs = 200)
```

```python
test_loss, test_acc = model.evaluate(data_set.iloc[:, 1:5], data_set.Species)
print('Test loss:', test_loss)
print('Test accuracy:', test_acc)
```
